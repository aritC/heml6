### This Notebook is used to check and merge the files successfully and remove NA values if any.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bs = pd.read_csv("og files/Building_Structure.csv")
```

```python
bs.head()
```

```python
bo = pd.read_csv("og files/Building_Ownership_Use.csv")
```

```python
bs.info()
```

```python
bo.info()
```

```python
train = pd.read_csv("og files/train.csv")
```

```python
test = pd.read_csv("og files/test.csv")
```

```python
result = pd.merge(bs,bo,on='building_id')
```

```python
result.head()
```

```python
res_train = pd.merge(train,result,on='building_id')
```

```python
res_test = pd.merge(test,result,on='building_id')
```

```python
res_train.info()
```

```python
res_test.info()
```

```python
res_train.to_csv('merged_train.csv')
```

```python
res_test.to_csv('merged_test.csv')
```

```python
plt.figure(figsize = (30 , 30))
sns.heatmap(res_train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
plt.figure(figsize = (30 , 30))
sns.heatmap(res_test.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

##### From the heatmap it is visible that only 'has_repair_started' column has NA values so the na values were replaced by 0 if mean was < 0.5 and 1 if mean was > 0.5.

```python
res_train['has_repair_started'].fillna(res_train['has_repair_started'].mean().round(),inplace=True)
```

```python
res_test['has_repair_started'].fillna(res_test['has_repair_started'].mean().round(),inplace=True)
```

#### Both the dataframes are saved so if we wanted to use any other technique to replace NA value we can do it and will not have to merge the files again.

```python
res_train.to_csv('nona_train.csv')
```

```python
res_test.to_csv('nona_test.csv')
```
